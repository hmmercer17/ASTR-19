```python
import numpy as np
from astropy.table import Table
```

This function sinx will use input values of x and return output values of sin(x)

```python
def sin_values(x):
    return np.sin(x)
```

This function cosx will use input values of x and return output values of cos(x)

```python
def cos_values(x):
    return np.cos(x)
```

This cell contains my variable 'x', which I have defined as 1000 evenly-spaced numbers between 0 and 2pi, as well as the table 't' that compares values of x, sin(x) and cos(x).

```python
x = np.linspace(0, 2*np.pi, 1000)

t = Table([x, sin_values(x), cos_values(x)], names=['x', 'sin(x)', 'cos(x)'])

#print(t)
```

This for loop will take my table 't' and print out only the first 10 values in each column.

```python
for i in range(10):
    print(t[i])
```
