```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
x = np.linspace(1,10,100)
y_data = a*np.exp(-b*x)

def model_1(x, a, b):
    return a*np.exp(-b*x)
```

```python
def model_2(x, a, b):
    return a/(x**b)
```

```python
a = 17.0
b = 21.2
```

```python
sigma = np.linspace(2, 0.2, len(x))
y_err = np.full(100, sigma)
scatter = np.random.normal(loc = 0, scale = sigma)

y_scatter = y_data + scatter
```

```python
params, params_cov = curve_fit(model_1, x, y_scatter, sigma=y_err)
m_fit1 = params[0]
b_fit1 = params[1]

params2, params_cov2 = curve_fit(model_2, x, y_scatter, sigma=y_err)
m_fit2 = params2[0]
b_fit2 = params[1]
print(f'The best fit values for Model 1 are m={m_fit1:4.3} and b={b_fit1:4.3}')
print(f'The best fit values for Model 2 are m={m_fit2:4.3} and b={b_fit2:4.3}')
```

```python
y_fit1 = model_1(x, m_fit1, b_fit1)

y_fit2 = model_2(x, m_fit2, b_fit2)
```

```python
plt.errorbar(x, y_scatter, yerr=y_err, fmt = 'o', label='Data', c='slategrey')
plt.plot(x, y_fit1, label = f'Model 1: m={m_fit1:3.2}, b={b_fit1:3.2}', c='indigo')
plt.plot(x, y_fit2, label = f'Model 2: m={m_fit2:3.2}, b={b_fit2:3.2}', c='cyan')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(fontsize=7)

plt.savefig('journal2_prompt3.pdf')
plt.show()
```
