```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,100)

def func(x, a, b):
    return a*np.exp(-b*x)

a = 4
b = 2.7
```

```python
y = func(x, a, b)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('$y = 4e^{-2.7x}$')

plt.savefig('journal2_prompt1_figure.pdf')
plt.show() 
```

### Start of Prompt 2

```python
sigma = np.linspace(2, 0.2, len(x))

scatter = np.random.normal(loc = 0, scale = sigma)

y_scatter = y + scatter

y_err = np.full(100, sigma)

plt.figure(figsize = (8,5))
plt.errorbar(x, y_scatter, yerr = y_err ,fmt = '*', c = 'g')

plt.xlabel('x')
plt.ylabel('y with scatter')
```
