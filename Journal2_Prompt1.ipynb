```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,100)

def func(x, a, b):
    return a*np.exp(-b*x)

a = 4
b = 2.7
```

```python
y = func(x, a, b)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('$y = 4e^{-2.7x}$')
```
